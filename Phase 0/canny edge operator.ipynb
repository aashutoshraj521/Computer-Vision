```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Computer Vision/karna.jpg", 0)

Gaus = cv2.GaussianBlur(image, (5,5), 0)

# Apply Canny
edges1 = cv2.Canny(
    Gaus,
    threshold1=50,
    threshold2=100
)
edges2 = cv2.Canny(
    Gaus,
    threshold1=100,
    threshold2=200
)
edges3 = cv2.Canny(
    Gaus,
    threshold1=200,
    threshold2=300
)

# Display
plt.figure(figsize=(10,4))

plt.subplot(2,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(edges1, cmap='gray')
plt.title("Canny Edges with 50 and 100 threshold")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(edges2, cmap='gray')
plt.title("Canny Edges with 100 and 200 threshold")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(edges3, cmap='gray')
plt.title("Canny Edges with 200 and 300 threshold")
plt.axis("off")

plt.tight_layout()
plt.show()
```
