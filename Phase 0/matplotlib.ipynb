```python
# Artificial Image Creation

import matplotlib.pyplot as plt
import numpy as np

image = np.array([
    [0, 80, 160],
    [60, 180, 255],
    [30, 120, 210]
])
plt.figure(figsize=(6, 20))

# With cmap="gray"
plt.subplot(1, 2, 1)
plt.imshow(image, cmap="gray")
plt.title("With cmap='gray'")

# Without cmap="gray" (Default)
plt.subplot(1, 2, 2)
plt.imshow(image)
plt.title("Without cmap='gray' (Viridis)")

plt.show()
```

```python
# Random Image Creation

import matplotlib.pyplot as plt
import numpy as np

image = np.random.randint(0, 256, (5,5))
print(image)
plt.figure(figsize=(4,4))
plt.imshow(image, cmap="gray")
plt.title("Random Image")
plt.axis("off")
plt.show()
```

```python
# Sub Plot

import matplotlib.pyplot as plt
import numpy as np

image = np.random.randint(0,256,(50,50))

plt.subplot(1,2,1)
plt.imshow(image,cmap="gray")
plt.title("Original")
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(image,cmap="gray")
plt.title("Copy")
plt.axis("off")

plt.show()
```

```python
# Plotting a Graph / Figure Customization
import matplotlib.pyplot as plt

epochs = [1,2,3,4,5]

loss = [2.3,1.8,1.2,0.8,0.4]

plt.figure(figsize=(6,4))

plt.plot(epochs, loss, label="Training Loss")

plt.title("Training Loss")

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.grid()

plt.legend()

plt.show()
```

```python
# Plotting a Histogram
import matplotlib.pyplot as plt
import numpy as np

image = np.random.randint(0,256,(100,100))

plt.hist(image.ravel(), bins=256)

plt.show()
```
