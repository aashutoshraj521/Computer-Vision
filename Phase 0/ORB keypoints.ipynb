```python
import cv2
import matplotlib.pyplot as plt

# 1 Read image
image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Images/chessboard.webp')

# 2 Convert to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 3 Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 4 Create ORB detector
orb = cv2.ORB_create(
    nfeatures=200
)

# 5 Detect keypoints and descriptors
keypoints, descriptors = orb.detectAndCompute(
    gray,
    None
)

# 6 Print:
# Number of keypoints
print(f'Number of keypoints: {len(keypoints)}')

# Keypoints
print(keypoints)

# Descriptor shape
print(descriptors.shape)

# 7 Draw keypoints
output = cv2.drawKeypoints(
    image_rgb,
    keypoints,
    None,
    color=(0,255,0),
    flags=0
)

# 8 Display image
plt.imshow(output)
plt.axis("off")
plt.show()
```
