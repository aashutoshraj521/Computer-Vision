```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Computer Vision/karna.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

horizontal = cv2.flip(image, 1)
vertical = cv2.flip(image, 0)
both = cv2.flip(image, -1)

plt.figure(figsize=(7.5,6))

plt.subplot(2,2,1)
plt.imshow(image)
plt.title("Original")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(horizontal)
plt.title("Horizontal")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(vertical)
plt.title("Vertical")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(both)
plt.title("Both")
plt.axis("off")

plt.tight_layout()
plt.show()
```
