```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Computer Vision/karna.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
blur1 = cv2.GaussianBlur(image, (5,5), 0)
blur2 = cv2.blur(image, (5,5))
blur3 = cv2.medianBlur(image, 5)
blur4 = cv2.bilateralFilter(image, 9, 75, 75)
```

```python
plt.subplot(2,3,1)
plt.title("Original")
plt.axis("off")
plt.imshow(image)

plt.subplot(2,3,2)
plt.title("Gaussian Blur")
plt.axis("off")
plt.imshow(blur1)

plt.subplot(2,3,3)
plt.title("Blur")
plt.axis("off")
plt.imshow(blur2)

plt.subplot(2,3,4)
plt.title("Median Blur")
plt.axis("off")
plt.imshow(blur3)

plt.subplot(2,3,5)
plt.title("Bilateral Filter")
plt.axis("off")
plt.imshow(blur4)

plt.tight_layout()
plt.show()
```
