```python
import cv2
import matplotlib.pyplot as plt

# Read image
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Computer Vision/karna.jpg")

# Convert BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Normalize pixel values to [0,1]
normalized = image_rgb / 255.0
# normalized = image_rgb - 100

# Display
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(image_rgb)
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(normalized)
plt.title("Normalized")

plt.show()
```

```python
print(normalized.min(), normalized.max())
```
