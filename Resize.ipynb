```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Computer Vision/karna.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

resized = cv2.resize(image, (224, 224))

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(resized)
plt.title("Resized")

plt.show()
```
