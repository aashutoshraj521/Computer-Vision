```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Computer Vision/karna.jpg", 0)

# Sobel gradients
gx = cv2.Sobel(image, cv2.CV_64F, 1, 0)
gy = cv2.Sobel(image, cv2.CV_64F, 0, 1)

# Gradient magnitude
magnitude = np.sqrt(gx**2 + gy**2)

plt.imshow(magnitude, cmap="gray")
plt.title("Before NMS")
plt.axis("off")
plt.show()
```

```python
# Apply Canny
edges = cv2.Canny(
    image,
    threshold1=100,
    threshold2=200
)

# Display
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(edges, cmap='gray')
plt.title("Canny Edges")
plt.axis("off")

plt.show()
```
