```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Computer Vision/Copy of mahadev.jpg")
print(type(image))
print(image.shape)
print(image.dtype)
print(image[100,100])
plt.imshow(image)
plt.axis("off")
cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/Computer Vision/Learning Tools for Research/copy.jpg', image)
```

```python
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
plt.axis("off")
cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/Computer Vision/Learning Tools for Research/copy.jpg', rgb)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap="gray")
plt.axis("off")
```
