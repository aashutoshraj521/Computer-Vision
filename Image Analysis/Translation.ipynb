```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Computer Vision/karna.jpg")

# Convert BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Translation amounts
tx = 100
ty = 50

# Translation matrix
T = np.float32([
    [1, 0, tx],
    [0, 1, ty]
])

# Apply translation
translated = cv2.warpAffine(
    image_rgb,
    T,
    (image_rgb.shape[1], image_rgb.shape[0])
)

# Display results
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(image_rgb)
plt.title("Original")
plt.axis("on")

plt.subplot(1,2,2)
plt.imshow(translated)
plt.title("Translated")
plt.axis("off")

plt.show()
```
