```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Computer Vision/karna.jpg")

# Convert BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Translation amounts
tx = 100
ty = 100

# Translation matrix
T1 = np.float32([
    [1, 0, tx],
    [0, 1, 0]
])
T2 = np.float32([
    [1, 0, 0],
    [0, 1, ty]
])

# Combine translation matrices
T = np.float32([
    [1, 0, tx],
    [0, 1, ty]
])

# Apply translation
translated1 = cv2.warpAffine(
    image_rgb,
    T1,
    (image_rgb.shape[1], image_rgb.shape[0])
)
translated2 = cv2.warpAffine(
    image_rgb,
    T2,
    (image_rgb.shape[1], image_rgb.shape[0])
)
translated = cv2.warpAffine(
    image_rgb,
    T,
    (image_rgb.shape[1], image_rgb.shape[0])
)

# Display results
plt.figure(figsize=(6,5))

plt.subplot(2,2,1)
plt.imshow(image_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(translated1)
plt.title("Horizontal Translation")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(translated2)
plt.title("Vertical Translation")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(translated)
plt.title("Combined Translation")
plt.axis("off")

plt.tight_layout()
plt.show()
```
