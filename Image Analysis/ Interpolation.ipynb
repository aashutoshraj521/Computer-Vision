```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Computer Vision/karna.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(3,3))
plt.axis("off")
plt.imshow(image)
print(image.shape)
```

```python
nearest = cv2.resize(image, (400,400),interpolation=cv2.INTER_NEAREST)
linear = cv2.resize(image, (400,400),interpolation=cv2.INTER_LINEAR)
cubic = cv2.resize(image, (400,400),interpolation=cv2.INTER_CUBIC)
area = cv2.resize(image, (400,400),interpolation=cv2.INTER_AREA)
```

```python
plt.figure(figsize=(3,3))
plt.subplot(2,2,1)
plt.axis("off")
plt.title("Nearest")
plt.imshow(nearest)
plt.subplot(2,2,2)
plt.axis("off")
plt.title("Linear")
plt.imshow(linear)
plt.subplot(2,2,3)
plt.axis("off")
plt.title("Cubic")
plt.imshow(cubic)
plt.subplot(2,2,4)
plt.axis("off")
plt.title("Area")
plt.imshow(area)
plt.tight_layout() # This prevents subplot titles from overlapping.
plt.show()
```

```python
print("Nearest ", nearest.shape)
print("Linear ", linear.shape)
print("Cubic ", cubic.shape)
print("Area ", area.shape)
```
