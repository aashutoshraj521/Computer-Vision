```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Computer Vision/coins.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

_, binary = cv2.threshold(
    gray,
    127,
    255,
    cv2.THRESH_BINARY
)

kernel = np.ones((3, 3), np.uint8)

eroded = cv2.erode(binary, kernel, iterations=1)
dilated = cv2.dilate(binary, kernel, iterations=1)
opened = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)
closed = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel)
gradient = cv2.morphologyEx(binary, cv2.MORPH_GRADIENT, kernel)

figure = plt.figure(figsize=(8, 6))
figure.add_subplot(2, 3, 1)
plt.imshow(binary, cmap='gray')
plt.title('binary')

figure.add_subplot(2, 3, 2)
plt.imshow(eroded, cmap='gray')
plt.title('eroded')

figure.add_subplot(2, 3, 3)
plt.imshow(dilated, cmap='gray')
plt.title('dilated')

figure.add_subplot(2, 3, 4)
plt.imshow(opened, cmap='gray')
plt.title('opened')

figure.add_subplot(2, 3, 5)
plt.imshow(closed, cmap='gray')
plt.title('closed')

figure.add_subplot(2, 3, 6)
plt.imshow(gradient, cmap='gray')
plt.title('gradient')

plt.tight_layout()
plt.show()
```
