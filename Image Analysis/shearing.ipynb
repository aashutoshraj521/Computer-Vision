```python
def display():
  plt.figure(figsize=(10,5))

  plt.subplot(1,2,1)
  plt.imshow(image_rgb)
  plt.title("Original")
  plt.axis("off")

  plt.subplot(1,2,2)
  plt.imshow(sheared)
  plt.title("Sheared")
  plt.axis("off")

  plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Computer Vision/karna.jpg")

# Convert BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Image dimensions
h, w = image_rgb.shape[:2]

# Shear factor
shear_factor = 0.5

# Shearing matrix
M = np.float32([
    [1, shear_factor, 0],
    [0, 1, 0]
])

# New width after shearing
new_width = int(w + abs(shear_factor) * h)

# Apply shearing
sheared = cv2.warpAffine(
    image_rgb,
    M,
    (new_width, h)
)

# Display
display()
```

```python
new_height = int(h + abs(shear_factor) * w)
M = np.float32([
    [1, 0, 0],
    [shear_factor, 1, 0]
])
sheared = cv2.warpAffine(
    image_rgb,
    M,
    (w, new_height)
)
# Display
display()
```

```python
M = np.float32([
    [1, shear_factor, 0],
    [shear_factor, 1, 0]
])
sheared = cv2.warpAffine(
    image_rgb,
    M,
    (new_width, new_height)
)
# Display
display()
```
