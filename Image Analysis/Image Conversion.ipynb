```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Computer Vision/Ram.jpg")
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
print("BGR Shape :", image.shape)
print("RGB Shape :", rgb.shape)
print("Gray Shape:", gray.shape)
print("HSV Shape :", hsv.shape)
```

```python
plt.subplot(2,2,1)
plt.title("Original BGR")
plt.imshow(image)
plt.axis("off")

plt.subplot(2,2,2)
plt.title("RGB Image")
plt.imshow(rgb)
plt.axis("off")

plt.subplot(2,2,3)
plt.title("Gray Image")
plt.imshow(gray, cmap="gray")
plt.axis("off")

plt.subplot(2,2,4)
plt.title("HSV Image")
plt.imshow(hsv)
plt.axis("off")

plt.show()
```
