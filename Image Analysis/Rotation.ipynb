```python
import cv2
import matplotlib.pyplot as plt

# Read image
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Computer Vision/karna.jpg")

# Convert BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Get image dimensions
(h, w) = image_rgb.shape[:2]

# Center of rotation
center = (w // 2, h // 2)

# Rotation matrix
M1 = cv2.getRotationMatrix2D(center, 45, 1.0)
M2 = cv2.getRotationMatrix2D(center, 90, 1.0)
M3 = cv2.getRotationMatrix2D(center, -30, 1.0)

# Rotate image
rotated1 = cv2.warpAffine(image_rgb, M1, (w+100, h+200))
rotated2 = cv2.warpAffine(image_rgb, M2, (w+100, h+200))
rotated3 = cv2.warpAffine(image_rgb, M3, (w+100, h+200))

# Display
plt.figure(figsize=(6,5))

plt.subplot(2,2,1)
plt.imshow(image_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(rotated1)
plt.title("Rotated 45°")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(rotated2)
plt.title("Rotated 90°")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(rotated3)
plt.title("Rotated -30°")
plt.axis("off")

plt.tight_layout()
plt.show()

print("Original Image Shape ", image_rgb.shape)
print("Rotated 45° Image Shape ", rotated1.shape)
print("Rotated 90° Image Shape ", rotated2.shape)
print("Rotated -30° Image Shape ", rotated3.shape)
```
