```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read grayscale image
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Computer Vision/Ram.jpg", 0)

# Scharr X
scharr_x = cv2.Scharr(
    image,
    cv2.CV_64F,
    1,
    0
)

# Scharr Y
scharr_y = cv2.Scharr(
    image,
    cv2.CV_64F,
    0,
    1
)
# There is no ksize like sobel has. It is because this operator is specially designed for 3*3 filter
# Gradient magnitude
scharr = np.sqrt(
    scharr_x**2 +
    scharr_y**2
)

# Display
plt.figure(figsize=(12,4))

plt.subplot(1,4,1)
plt.imshow(image, cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1,4,2)
plt.imshow(np.abs(scharr_x), cmap='gray')
plt.title("Scharr X")
plt.axis("off")

plt.subplot(1,4,3)
plt.imshow(np.abs(scharr_y), cmap='gray')
plt.title("Scharr Y")
plt.axis("off")

plt.subplot(1,4,4)
plt.imshow(scharr, cmap='gray')
plt.title("Combined")
plt.axis("off")

plt.show()
```
