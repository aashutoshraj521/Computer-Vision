```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Computer Vision/Ram.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print(image.shape)
crop = image[200:800, 250:750]
plt.figure(figsize=(4,2))

plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(crop)
plt.title("Crop")
plt.axis("off")

plt.show()
```

```python
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Computer Vision/karna.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print(image.shape)

h, w = image.shape[:2]

lc = image[0:200, 0:200]
print("Left Corner Shape ", lc.shape)

center = image[h//2-100:h//2+100, w//2-100:w//2+100]
print("Center Region ", center.shape)

rc = image[h-200:h, w-200:w]
print("Right Corner ", rc.shape)

plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(lc)
plt.title("Left Corner")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(center)
plt.title("Center Region")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(rc)
plt.title("Right Corner")
plt.axis("off")

plt.tight_layout()
plt.show()
```
