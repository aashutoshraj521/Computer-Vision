```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Computer Vision/Ram.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print(image.shape)
crop = image[200:800, 250:750]
plt.figure(figsize=(4,2))

plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(crop)
plt.title("Crop")
plt.axis("off")

plt.show()
```

```python
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Computer Vision/karna.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print(image.shape)

lc = image[0:100, 0:100]
print("Left Corner Shape ", lc.shape)
cr = image[400:600, 300:400]
print("Center Region ", cr.shape)
rc = image[900:1061, 700:750]
print("Right Corner ", rc.shape)

plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(lc)
plt.title("Left Corner")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(cr)
plt.title("Center Region")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(rc)
plt.title("Right Corner")
plt.axis("off")

plt.tight_layout()
plt.show()
```
