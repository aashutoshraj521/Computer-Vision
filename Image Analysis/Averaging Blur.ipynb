```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Read image
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Computer Vision/Copy of mahadev.jpg")

# Averaging blur
blurred = cv2.blur(
    image,
    (7,7)
)

# Resize for comparison
size = (300,300)

original_r = cv2.resize(image, size)
blurred_r = cv2.resize(blurred, size)

comparison = np.hstack([
    original_r,
    blurred_r
])

cv2_imshow(comparison)
```
